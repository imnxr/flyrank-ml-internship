## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

```python
# The Question

"""
->Research Question

How can machine learning help identify which content pages should be prioritized for optimization using search performance and engagement signals?

->Unit of Analysis

One content page.

->Output

A priority score or recommendation indicating whether a page should be refreshed or investigated.

->Decision

An SEO or content manager decides which pages should receive attention first.

->Action

The team reviews and updates the highest-priority pages instead of manually checking every page.

->Cost of a Wrong Recommendation

A false positive may waste time and resources updating pages that do not need changes. 
A false negative may cause important pages with declining performance to be ignored, resulting in missed traffic or engagement opportunities. 
The model is intended to support human decisions rather than replace them."""
```

## 3. Quick look at the data (2-3 real numbers)
